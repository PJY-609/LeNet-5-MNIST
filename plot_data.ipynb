```python
metric = list(baseline.columns)
print(metric)
```

```python
print(baseline['epoch'].tolist())
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
epochs = list(baseline['epoch'])
for i, m in enumerate(metric):
    if m != 'epoch':
        if i == 1 or i == 3:
            loc = 0
        else:
            loc = 1
        plt.subplot(2, 2, i)
        line1, = plt.plot(epochs, baseline[m].tolist(), ls='-', linewidth=4, label='Basline')
        line2, = plt.plot(epochs, randnorm[m].tolist(), ls='--', linewidth=4, label='Random Normal')
        line3, = plt.plot(epochs, he_norm[m].tolist(), ls='-.', linewidth=4, label='He Normal')
        line4, = plt.plot(epochs, he_uni[m].tolist(), ls=':', linewidth=4, label='He Uniform')
        plt.legend(handles=[line1, line2, line3, line4], loc=loc, fontsize=15)
        plt.title(m, fontsize=20)
        
#         plt.subplot(2, 2, i+2)
#         line5, = plt.plot(epochs, baseline[m].tolist(), ls='-', linewidth=4, label='Basline')
#         line6, = plt.plot(epochs, RMSprop[m].tolist(), ls='--', linewidth=4, label='RMSprop')
#         line7, = plt.plot(epochs, momentum[m].tolist(), ls='-.', linewidth=4, label='Momentum')
#         line8, = plt.plot(epochs, Adam[m].tolist(), ls=':', linewidth=4, label='Adam')
#         plt.title(m, fontsize=20)

#         plt.legend(handles=[line5, line6, line7, line8], loc=loc, fontsize=15)
#         plt.title(alphabet[i-1], fontsize=20)

plt.savefig('3.jpg')
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
for i, m in enumerate(metric):
    if m != 'epoch':
        if i == 1 or i == 3:
            loc = 0
        else:
            loc = 1
        plt.subplot(2, 2, i)
        line5, = plt.plot(epochs, baseline[m].tolist(), ls='-', linewidth=4, label='Basline')
        line6, = plt.plot(epochs, RMSprop[m].tolist(), ls='--', linewidth=4, label='RMSprop')
        line7, = plt.plot(epochs, momentum[m].tolist(), ls='-.', linewidth=4, label='Momentum')
        line8, = plt.plot(epochs, Adam[m].tolist(), ls=':', linewidth=4, label='Adam')
        plt.title(m, fontsize=20)

        plt.legend(handles=[line5, line6, line7, line8], loc=loc, fontsize=15)
        plt.title(m, fontsize=20)

plt.savefig('3.jpg')
plt.show()
```
